```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
train_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
train_df
```

```python
test_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
test_df
```

```python
train_df['SalePrice'].mean()
```

```python
test_df['SalePrice'] = pd.Series([train_df['SalePrice'].mean()] * len(train_df))
test_df
```

```python
test_df[['Id', 'SalePrice']].to_csv('average prediction.csv', index=False)
```

```python
from copy import deepcopy 

train_df_sub = deepcopy(train_df[['MSSubClass', 'LotFrontage', 'SalePrice']])

train_df_sub
```

```python
train_df_sub = train_df_sub.fillna(train_df_sub.mean())
train_df_sub
```

```python
x_train, y_train = train_df_sub.to_numpy()[:, :-1], train_df_sub.to_numpy()[:, -1]

x_train 
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression().fit(x_train, y_train)
lr
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
88327 + 60*76 + 65*1260 
```

```python
train_df_sub['linear_prediction']= lr.predict(x_train)
train_df_sub
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(train_df_sub['SalePrice'], train_df_sub['linear_prediction'])
```

```python
test_df_sub = deepcopy(test_df[['MSSubClass', 'LotFrontage']])

test_df_sub
```

```python
test_df_sub = test_df_sub.fillna(test_df_sub.mean())
test_df_sub
```

```python
x_test = test_df_sub.to_numpy()
x_test
```

```python
test_df['SalePrice'] = lr.predict(x_test)
test_df
```

```python
test_df[['Id', 'SalePrice']].to_csv('Linear prediction.csv', index=False)
```
